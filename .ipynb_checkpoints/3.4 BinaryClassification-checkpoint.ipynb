```python
#Two-class classification, or binary classification, is a widely applied machine learning problem
#In this example, we will learn how to classify movie review as positive or negative, based on the review text

from keras.datasets import imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words = 10000)
#num_words = 10000 means that we will only use the most frequent 10000 words in the training data. We ignore infrequent words.
```

```python
#Variables train_data and test_data are list of word indices of words in reviews.
#Variables train_lables and test lables are list of 0, which indicates negative, and 1, which indicates positive
print(train_data[0])
print(train_labels[0])

#Let's see how we convert the review data back to english
word_index = imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in train_data[0]])
print(decoded_review)
```

```python
"""
We cannot input number list in neural network. There are two ways to convert list to tensor:
1. Add padding to the list so that all of the review list has same length, and convert the list into tensor with shape 
(samples, sequence_length). Then, use the layer that can handle this integer tensor as a first layer of neural network. (Embedding)
2. Apply one-hot encoding to the list to convert it into vector of 0 and 1. For example, convert [3, 5] as a vector whose values of 
index 3 and 5 are 1 and values of other indices are all 0. Then, use the Dense layer that can handle floating point vector data as a
first layer of neural network.
"""
#Using one-hot encoding
import numpy as np
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
print(x_train)

y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

```python
"""
Input data is vector and label is scalar. Network that works well on this kind of problem is stack of fully connected layer which uses
relu activation function (Dense(16, activation='relu'))
Argument in the Dense layer (16) is number of hidden unit. Hidden unit can be interpreted as a degree of freedom that neural network can
get when it is in process of training. For example, 16 hidden units means that weight tensor W has the shape of 
(input_dimension, 16). When dot operation is done to input data and W, input data is projected to a space that is expressed as 
16-dimension (simply, dimension of first axis changes from input_dimension to 16 with some calculations). If we increase the number of
hidden units, while the neural network would be able to learn more complicated expressions, but its computing cost will increase and it
may learn unexpected pattern (which helps on training data but not on test data).

Two things needs to be considered when stacking Dense layers:
1. How many layers will I use?
2. How many hidden units will each layer have?

We will learn how to decide these later. For now we will use:
1. 2 hidden layer with 16 hidden units
2. Third layer that outputs review's rating as prediction of scalar value.

Hidden layer in the middle uses relu as its activation function, and the last layer uses sigmoid activation function to output 
probability(a score between 0 and 1, where close to 0 means likely to be negative and close to 1 means likely to be positive). Relu is
a function that makes negative values 0. Sigmoid is a function that compresses some number into a range of [0, 1], meaning that we can
interpret its output as a probability.
"""

#Plotting relu function
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 100)
y = np.maximum(x, 0)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, y)
plt.show()

#Plotting sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
y = sigmoid(x)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, y)
plt.show()

#Implementation of this neural network
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

"""
What is activation function, and why do we need it?
Suppose that there is no activation function (or non-linearity) like relu. Then Dense layer will consist of dot operation and addition,
which are both linear operations (Remember linear algebra? Both dot operation and addition preserves linearity)
output = dot(W, input) + b.
However, this only enables neural network to learn only linear transformation over input. Also, stacking up layers just end up as a
simple linear transformation, so there is no benefit on stacking multiple layers. For example, suppose that there are three layers
L1 = dot(W1, input) + b1, L2 = dot(W2, input) + b2, and L3 = dot(W3, input) + b3. If we stack up these layers, the output will be
output = dot(W3, (dot(W2, (dot(W1, input) + b1)) + b2)) + b3. However, since all of them are linear operations which preserves 
linearity, we can also write this as output = dot(W4, input) + b4. Observe that this final expression is no different from just one
layer.

Therefore, to benefit from stacking up layers and making hypothesis space expressive, we need activation function (or non-linearity).
Relu is the most popular activation function in deep learning.
"""
```

```python
"""
Lastly, we need to select loss function and optimizer. Since this is a binary classification problem and the output of the neural 
network is probability, binary_crossentropy loss function is suitable, although it is not the only option (mean_squared_error can be
used). Crossentropy is the best option when using model that outputs probability. It measures difference between probability 
distributions. Here, it measures difference between true distribution and predicted distribution.
"""

#Setting the model with rmsprop optimizer and binary_crossentropy loss function. We will monitor the training process using accuracy.
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

"""
Since rmsprop, binary_crossentropy, and accuracy are included in keras, we can select these using string literal. Sometimes, we may
have to change parameter of the optimizer or loss function, metrics of our own. Former can be done by creating optimizer python object
and pass it to the optimizer parameter, and latter can be done by passing functor to loss or metrics parameter.

Example of setting optimizer on our own:
from keras import optimizers
model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])

Example of assigning loss and metrics as functors:
from keras import losses
from keras import metrics
model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=[metrics.binary_accuracy])
"""
```

```python
"""
In order to measure the accuracy about the data first seen by model, you must create validation set by taking 10000 samples from 
original training data.
"""

x_val = x_train[:10000]
partial_x_train = x_train[10000:]
y_val = y_train[:10000]
partial_y_train = y_train[10000:]

"""
Now we will train model for 20 epochs by creating mini batches that consists of 512 samples (repeat training process 20 times for all
samples in x_train and y_train tensors). Simultaneosly, we will measure loss and accuracy for 10000 samples we haven taken from original
training data before. To do this, we have to pass validation set to validation_data parameter.
"""

#Training the model
history = model.fit(partial_x_train, partial_y_train, epochs=20, batch_size=512, validation_data=(x_val, y_val))

"""
model.fit() method returns History object. This object has history attribute which is a dictionary that holds all information gathered 
from the process of training.
"""

#We can use History object returned from model.fit() to plot a graph of training and validation loss
history_dict = history.history
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)

plt.clf()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

#Plotting graph of training and validation accuracy
plt.clf()
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
"""
Above, we can see that training loss decrease per epoch, and training accuracy increase per epoch. This is expected since repetition of 
stochastic gradient descent decreases loss per repetition. However, validation loss does not show this pattern. It seems that graph's
direction changes at 4th epoch. This is one case of model working well on training set not working well on first-seen data.
Specifically, it is said to be overfitting. After second epoch, the model is too fit to the training data so it learns expression
specialized to the training data which cannot be generalized to data other than training set. In this case, we can halt the training
after about 3 epochs to prevent overfitting. Later we will learn about skills to prevent overfitting.
"""

#Neural network training for only 4 epochs
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=4, batch_size=512)
results = model.evaluate(x_test, y_test)
print(results)
#We have achieved accuracy of 88% with simple procedure. We can get 95% if we use better techniques
```

```python
"""
We would want to try our model in real life after training. We can use predict() method to predict the probability that some review is
positive.
"""
print(model.predict(x_test))

"""
Like this, this model have confidence on some samples (>0.99 or <0.01), but lacks confidence on other samples (0.6, 0.4, etc)
"""
```
