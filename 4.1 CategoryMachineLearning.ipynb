```python
"""
In previous examples, we have handled three types of machine learning problems: binary classification, multiclass classification, and
scalar regression. These are all examples of supervised learning. Objective of supervised learning is learning relationship between 
input and target of training data. But supervised learning is just one part of machine learning. There are mainly four categories in
machine learning:
1. Supervised learning: It learns how to map input data to known target data given sample data (often labeled by people). Optical 
character recognition, voice recognition, image classification, and language translation belong to this category. Supervised learning
usually consists of classification and regression, but there are some other variety:
- Sequence generation: Create caption that explains the given picture. It sometimes can be reconstructed to sequence of classification
problem (like repeatedly predicting word or token in the sequence).
- Syntax tree prediction: Predict syntax tree given a sentence
- Object detection: Draw bounding area around some target object in a given picture. It can be expressed as a classification problem
(that classifies contents of each box given many bounding area candidates), or expressed as a combination of regression and 
classification that predicts coordinates of bounding area by vector regression.
- Image segmentation: Do masking on some target object by pixel given a picture

2. Unsupervised learning: This type of machine learning does not use any target and finds interesting conversion about input data. We 
use it for data visualization, data compression, data noise removal, and for better understanding of correlation in data. Unsupervised
learning is an important element in data analysis, and it is a necessary step for better understanding of dataset before solving 
supervised learning problem. Dimensionality reduction and clustering are well known categories in unsupervised learning.

3. Self-supervised learning: This is a special case of supervised learning but it is different enough to be a sepearate category. It is
a supervised learning that does not use labels that people created. We can think it as a supervised learning without human intervention.
Label is still needed because learning should be supervised in right direction, but it generates them from input data usually by using 
heuristic algorithms. For example, autoencoder is a well known example of self-supervised learning. In here generated target is original
input. In the same way predicting next frame of the video given previous frame or predicting next word given previous word are examples
of self-supervised learning (we call them temporally supervised learning since model is supervised from future input data). The boundary
between supervised, unsupervised, and self-supervised learning can sometimes be vague. They are continuous without sppecific boundary.
Self-supervised learning can be interpreted as supervised or unsupervised learning depending on where we focus on between learning 
mechanism or application of it.

4. Reinforcement learning: it started to get a lot of focus after Google DeepMind was successfully applied on learning Atari gameplay. 
In reinforcement learning, agent is learned to choose action that maximizes the reward after getting information about the environment. 
For example, neural network that was trained by reinforcement learning can take video game screen as its input and output action to take
in game to maximize the game score. Currently reinforcement learning is in process of research and there is no success story beyond 
game. However, when the time comes it is expected that reinforcement learning will replace many real-world applications. These
applications include self-driving car, resource management, and education.

There are many vocabularies in classification and regression:
- Sample or input: one data point that will be inserted into model
- Prediction or output: value that comes out of the model
- Target: the correct answer; the value that model needs to perfectly predict based on external data source
- Forecast error or loss value: value that measures the distance between model's prediction and target
- Class: set of labels that can be chosen in classification problem; for example if we are classifying pictures of cats and dogs, class
of them is 'dog' and 'cat'
- Label: specific case of class assignment in classification problem; for example if we denote that 'dog' class is in picture #1234,
'dog' becomes a label of picture #1234
- Ground-truth or annotation: all targets about the dataset; it is usually collected by human
- Binary classification: classification task where each input sample is classified by two exclusive categories
- Multiclass classification: classification task where each input sample is classified by more than two exclusive categories
- Multilabel classification: classification task where each input sample can be assign to more than one label; for example if there are
both cat and dog in an image, we need to assign both 'cat' and 'dog' labels to the image; usually number of labels assigned differ by
image
- Scalar regression: task where target is continuous scalar value; different targets generate continuous space
- Vector regression: task where target is set of continuous scalar values; it is vector regression if it regresses about multiple values
- Mini-batch or batch: small bundle of samples that are processed at the same time by model; it is common to set their number to be a
power of 2 to make GPU's memory allocation easy; we calculate updated value of stochastic gradient descent method per mini batch when
training
"""
```
